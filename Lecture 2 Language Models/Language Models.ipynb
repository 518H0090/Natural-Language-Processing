```python
import collections, nltk
from nltk import ngrams
nltk.download('punkt')
```

```python
filename = 'train.txt'
with open(filename,encoding="utf8") as f:
    lines_train = f.readlines()
    
filename = 'test.txt'
with open(filename,encoding="utf8") as f:
    lines_test = f.readlines()
```

```python
# X√¢y d·ª±ng m√¥ h√¨nh ng√¥n ng·ªØ 1-gram
def unigram(lines):
    unigram_dict={}
    total_count = 0
    for sent in lines:
        words = nltk.word_tokenize(sent)
        total_count += len(words)
        for w in words:
            if w in unigram_dict.keys():
                unigram_dict[w] += 1
            else:
                unigram_dict[w] = 1
    return unigram_dict
print(unigram(lines_train))
```

```python
# X√¢y d·ª±ng m√¥ h√¨nh ng√¥n ng·ªØ 2-gram
def bigram(lines):
    bigram_dict={}
    for sent in lines:
        words = nltk.word_tokenize(sent)
        for i in range(len(words)):
            if i==0:
                gram='<s>_'+words[i]
            else:
                gram=words[i-1]+'_'+words[i]

            if gram in bigram_dict.keys():
                bigram_dict[gram] += 1
            else:
                bigram_dict[gram] = 1
    return bigram_dict
print(bigram(lines_train))
```

```python
# X√¢y d·ª±ng m√¥ h√¨nh ng√¥n ng·ªØ 3-gram
def trigram(lines):
    trigram_dict={}
    for sent in lines:
        words = nltk.word_tokenize(sent)
        for i in range(len(words)):
            if i==0:
                gram='<s>_'+words[i]
            elif i == 1:
                gram=words[i-1]+'_'+words[i]
            else:
                gram=words[i-2]+'_'+words[i-1]+'_'+words[i]

            if gram in trigram_dict.keys():
                trigram_dict[gram] += 1
            else:
                trigram_dict[gram] = 1
    return trigram_dict
print(trigram(lines_train))
```
