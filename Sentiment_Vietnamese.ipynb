```python
import pandas as pd
import nltk 
import numpy as np
```

```python
!wget https://raw.githubusercontent.com/nthanhkhang/Natural-Language-Processing/main/Lecture%204%20Text%20classification/data/data%20-%20data.csv
```

```python
filename = 'data - data.csv'
df = pd.read_csv(filename)
print(df.head(5))
print(df.shape)
```

```python
data = df.values
print(data.shape)
```

```python
print(data[:10])
```

```python
lines = data[:,0]
labels = data[:,1]
print(lines[:20])
print(labels[:20])
```

```python
nltk.download('stopwords')
nltk.download('punkt')
from nltk.corpus import stopwords
def remove_stop_words(words):
    result = [i for i in words if i not in stopwords.words('english')]
    return result
```

```python
sentences = [nltk.word_tokenize(sent) for sent in lines]
print(sentences[0])
remove_stop_words(sentences)
```

```python
from keras.preprocessing.text import Tokenizer 
tokenizer = Tokenizer()
tokenizer.fit_on_texts(sentences)
```

```python
X = tokenizer.texts_to_matrix(sentences,mode='binary')
print(X)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, labels, test_size=0.2, random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression  
model_LR = LogisticRegression()
model_LR = model_LR.fit(X_train,y_train)
y_pred = model_LR.predict(X_test)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test,y_pred))
```

```python
from sklearn.naive_bayes import MultinomialNB   
model_NB = MultinomialNB()
model_NB = model_NB.fit(X_train,y_train)
y_pred = model_NB.predict(X_test)

print(classification_report(y_test,y_pred))
```

```python
input_sent = 'tất sờn hết ca chưa dùng mà vay r' 
tokens = nltk.word_tokenize(input_sent)
print(tokens)
X_input = tokenizer.texts_to_matrix([tokens],mode='binary')
print(X_input)
```

```python
y_pred = model_NB.predict(X_input)
print(y_pred)
```
